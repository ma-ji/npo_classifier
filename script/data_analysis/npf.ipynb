```python
import pandas as pd
```

```python
df_ein_remap_ntee_2018 = pd.read_excel("../../output/df_ein_remap_ntee_2018.xlsx")
hdr = {
    "User-Agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.11 (KHTML, like Gecko) Chrome/23.0.1271.64 Safari/537.11",
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8",
    "Accept-Charset": "ISO-8859-1,utf-8;q=0.7,*;q=0.3",
    "Accept-Encoding": "none",
    "Accept-Language": "en-US,en;q=0.8",
    "Connection": "keep-alive",
}
df_bmf_201812 = pd.read_csv(
    "https://nccs-data.urban.org/dl.php?f=bmf/2018/bmf.bm1812.csv", storage_options=hdr
)
```

```python
df_ein_remap_ntee_2018 = df_ein_remap_ntee_2018.merge(
    df_bmf_201812, on="EIN", how="left"
)
```

```python
df_ein_remap_ntee_2018
```

```python
df_ntee_c = df_ein_remap_ntee_2018[df_ein_remap_ntee_2018.Remap_NTEE == "C"][
    ["EIN", "Remap_NTEE", "NTEE1"]
]
df_ntee_c.drop_duplicates(subset=["EIN", "Remap_NTEE"], inplace=True)
# Show raw counts and by column percentages

df_c_stats = pd.concat(
    [
        df_ntee_c.NTEE1.value_counts(),
        # Show percentages with 2 decimal places and add a % sign
        df_ntee_c.NTEE1.value_counts(normalize=True).apply(
            lambda x: f"{round(x*100, 2)}%"
        ),
    ],
    axis=1,
)
# Add a total row

df_c_stats.loc["Total"] = [
    df_ntee_c.NTEE1.value_counts().sum(),
    df_ntee_c.NTEE1.value_counts(normalize=True).sum(),
]
df_c_stats
```

```python
df_c_stats.to_csv("../../output/df_c_stats.tsv", sep="\t")
```

```python
df_ntee_c_tx = df_ein_remap_ntee_2018[
    (df_ein_remap_ntee_2018.Remap_NTEE == "C") & (df_ein_remap_ntee_2018.STATE == "TX")
][["EIN", "Remap_NTEE", "NTEE1"]]
df_ntee_c_tx.drop_duplicates(subset=["EIN", "Remap_NTEE"], inplace=True)
# Show raw counts and by column percentages

df_c_stats_tx = pd.concat(
    [
        df_ntee_c_tx.NTEE1.value_counts(),
        # Show percentages with 2 decimal places and add a % sign
        df_ntee_c_tx.NTEE1.value_counts(normalize=True).apply(
            lambda x: f"{round(x*100, 2)}%"
        ),
    ],
    axis=1,
)
# Add a total row

df_c_stats_tx.loc["Total_TX"] = [
    df_ntee_c_tx.NTEE1.value_counts().sum(),
    df_ntee_c_tx.NTEE1.value_counts(normalize=True).sum(),
]
df_c_stats_tx
```

```python
df_c_stats_tx.to_csv("../../output/df_c_stats_tx.tsv", sep="\t")
```
